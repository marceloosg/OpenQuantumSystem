```python
from krotov import Krotov,KrotovParams,controller as ctrl,np 
import matplotlib.pyplot as plt
fig=None
ax=None
def init():
    fig, ax= plt.subplots()
    return ax

def legend():
    ax.legend(loc='upper right', frameon=False,shadow=True, fontsize='x-large',
              ncol =2,bbox_to_anchor=(2, 1))
def plot(rho, txt="rho"):
    txt0="{} diag real".format(txt)
    txt3="{} diag imag".format(txt)
    txt1="{} off real".format(txt)
    txt2="{} off imag".format(txt)
    ax.plot(np.real(rho.states[:,[0,3]]),label=txt0)
    ax.plot(np.imag(rho.states[:,[0,3]]),label=txt3)
    ax.plot(np.real(rho.states[:,[1,2]]), label=txt1)
    ax.plot(np.imag(rho.states[:,[1,2]]), label=txt2)
```

```python

parameters = KrotovParams(omega=4.)
H = ctrl.LiouvilleHamiltonian(parameters)

times_krotov = 25

k = Krotov(H, params=parameters)  #
k.run_krotov(times_krotov)
```
