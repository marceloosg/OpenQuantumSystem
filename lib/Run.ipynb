```python
from hamiltonian import Hamiltonian
from krotov import Krotov
import params as p
import numpy as np
```

```python
target = np.array([[0.4, 0], [0, 0.6]], dtype='complex')
# target[3] = 1/(1+exp(-Bw))

H = Hamiltonian(gamma=1., omega=2., target=target)
parameters = p.KrotovParams(eta=1.5, delta=1.5, alpha=0.001, num_t=2000, epsilon=0.01, speedup_factor=2)
initial_rho = np.array([[0.5, -0.19j], [0.19j, 0.5]], dtype='complex')
times_krotov = 50

k = Krotov(H, params=parameters, rho=initial_rho, thermal_target=target)  #
```

```python


k.run_krotov(times_krotov)  

```

```python
import matplotlib.pyplot as plt
plt.plot(range(times_krotov), np.array(k.cost), 'r')
```
